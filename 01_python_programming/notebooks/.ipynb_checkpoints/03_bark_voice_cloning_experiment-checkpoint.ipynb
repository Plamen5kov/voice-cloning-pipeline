```python
"""
Voice Cloning with Bark: Use a WAV sample as history prompt
"""
from bark import SAMPLE_RATE, generate_audio, load_history_prompt
from IPython.display import Audio, display

# Path to your processed WAV sample
wav_path = '../data/processed_voice_samples/quick-brown-fox.wav'

# Load the history prompt from your WAV file
history_prompt = load_history_prompt(wav_path)

# Text to synthesize in your cloned voice
text = "This is a test of voice cloning using my own sample."

# Generate audio using Bark with the history prompt
cloned_audio = generate_audio(text, history_prompt=history_prompt)
print("Cloned Voice Output:")
display(Audio(cloned_audio, rate=SAMPLE_RATE))
```
