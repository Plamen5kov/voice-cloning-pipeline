```python
"""
Notebook for feature extraction and Bark inference.
- Visualize spectrograms of preprocessed audio
- Run Bark TTS inference on sample text
"""
import os
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
from IPython.display import Audio, display

# Path to preprocessed audio
AUDIO_DIR = '../data/libritts_sample/preprocessed/'

# List audio files
files = [f for f in os.listdir(AUDIO_DIR) if f.endswith('.wav')]

# Visualize spectrograms
for fname in files:
    path = os.path.join(AUDIO_DIR, fname)
    y, sr = librosa.load(path, sr=None)
    plt.figure(figsize=(10, 4))
    S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=80)
    S_dB = librosa.power_to_db(S, ref=np.max)
    librosa.display.specshow(S_dB, sr=sr, x_axis='time', y_axis='mel')
    plt.title(f'Mel Spectrogram: {fname}')
    plt.colorbar(format='%+2.0f dB')
    plt.tight_layout()
    plt.show()
    display(Audio(y, rate=sr))

# Bark inference example (text-to-speech)
from bark import SAMPLE_RATE, generate_audio

text = "This is a test of Bark text to speech synthesis."

# Generate audio from text
bark_audio = generate_audio(text)

# Play generated audio
print("Bark TTS output:")
display(Audio(bark_audio, rate=SAMPLE_RATE))
```
