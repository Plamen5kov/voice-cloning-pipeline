```python
"""
Bark TTS experiment: Generate an orc-like voice
"""
from bark import SAMPLE_RATE, generate_audio
from IPython.display import Audio, display

orc_text = "You dare challenge me, human? Prepare to face the wrath of the orc horde!"
orc_prompt = "[man, deep] " + orc_text

orc_audio = generate_audio(orc_prompt)
print("Orc TTS output:")
display(Audio(orc_audio, rate=SAMPLE_RATE))
```
